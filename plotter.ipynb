```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# HDBSCAN, minPts = 2
categories = ['FS_1', 'FS_2']
ch_index = [1171.7446, 1868.9511]  # ch_index values
clusters = [115249, 116333]
noise_points = [316442, 316265]
# Creating the bar chart
fig, ax1 = plt.subplots()

# Creating bar width and position
bar_width = 0.1
index = np.arange(len(categories))

# Plotting the ch_index
color_v1 = '#06C'
bars1 = ax1.bar(index, ch_index, bar_width, color=color_v1, label='CH Index')
ax1.set_xlabel('Feature Sets')
ax1.set_ylabel('CH Index (higher is better)', color=color_v1)
ax1.tick_params('y', colors=color_v1)
ax1.set_ylim(0, 2000)  # Setting the maximum value to 2000
ax1.set_xticks(index)
ax1.set_xticklabels(categories)

ax2 = ax1.twinx()
color_v2 = '#38812F'
bars2 = ax2.bar(index+bar_width, clusters, bar_width, color=color_v2, label='Clusters')
ax2.set_ylabel('Clusters', color=color_v2)
ax2.tick_params('y', colors=color_v2)
ax2.set_ylim(110000, 120000)  # Setting the maximum value to 2000

ax3 = ax1.twinx()
color_v3 = '#7D1007'
bars3 = ax3.bar(index+2*bar_width, noise_points, bar_width, color=color_v3, label='Noise')
ax3.spines['right'].set_position(('outward', 60)) 
ax3.set_ylabel('Noise', color=color_v3)
ax3.tick_params('y', colors=color_v3)
ax3.set_ylim(316000, 317000)  # Setting the maximum value to 2000


# Adding a title

fig.legend([bars1, bars2, bars3], ['CH Index', 'Clusters', 'Noise'], loc='upper left', bbox_to_anchor=(0.12,0.88))

# Displaying the chart
plt.show()
```

```python

# Data for two experiments
categories = ['FS_1', 'FS_2']
one_subnet = [6378, 6688]  # ch_index values
three_subnets = [48882, 49314]
one_country = [52066, 52384]
more_3_countries = [884, 867]
# Creating the bar chart
fig, ax1 = plt.subplots()

# Creating bar width and position
bar_width = 0.1
index = np.arange(len(categories))

# Plotting the ch_index
color_v1 = '#06C'
bars1 = ax1.bar(index, one_subnet, bar_width, color=color_v1, label='1 subnet')
ax1.set_xlabel('Feature Sets')
ax1.set_ylabel('1 subnet', color=color_v1)
ax1.tick_params('y', colors=color_v1)
ax1.set_ylim(6000, 7000)  # Setting the maximum value to 2000
ax1.set_xticks(index)
ax1.set_xticklabels(categories)

ax2 = ax1.twinx()
color_v2 = '#4CB140'
bars2 = ax2.bar(index+bar_width, three_subnets, bar_width, color=color_v2, label='<=3 subnets')
ax2.set_ylabel('<=3 subnets', color=color_v2)
ax2.tick_params('y', colors=color_v2)
ax2.set_ylim(47500, 50000)  # Setting the maximum value to 2000

ax3 = ax1.twinx()
color_v3 = '#7D1007'
bars3 = ax3.bar(index+2*bar_width, one_country, bar_width, color=color_v3, label='Noise')
ax3.spines['right'].set_position(('outward', 60)) 
ax3.set_ylabel('>3 subnets, 1 country', color=color_v3)
ax3.tick_params('y', colors=color_v3)
ax3.set_ylim(51000, 53500)  # Setting the maximum value to 2000

ax4 = ax1.twinx()
color_v4 = '#005F60'
bars4 = ax4.bar(index+3*bar_width, more_3_countries, bar_width, color=color_v4, label='Noise')
ax4.spines['right'].set_position(('outward', 120)) 
ax4.set_ylabel('>3 countries', color=color_v4)
ax4.tick_params('y', colors=color_v4)
ax4.set_ylim(860, 900)  # Setting the maximum value to 2000

fig.legend([bars1, bars2, bars3, bars4], ['1 subnet', '<3 subnets', '>3 subnets, 1 country', '>3 countries'], loc='upper left', bbox_to_anchor=(0.12,0.88))

# Displaying the chart
plt.show()
```
